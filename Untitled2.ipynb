```
import tensorflow as tf
import numpy as np
```

```
tf.enable_eager_execution()
```

```
tf.__version__
```

```
X = np.array([
              [0,0],
              [0,1],
              [1,0],
              [1,1]],dtype='float32')
```

```
y = np.array( [
               [0],
               [0],
               [0],
               [1]],dtype='float32')
```

```
w = tf.Variable([[1],[2]],dtype='float32')
b = tf.Variable([3],dtype='float32')
```

```
tf.matmul(X,w)
```

```
tf.matmul(X,w)+b
```

```
optimizer = tf.train.AdamOptimizer(0.01)
```

```
optimizer
```

```
#Optimizer = w/b 를 업데이트
#tf.GradienTape() = w_gred / b_gred 의 값을 구해주기 
```

```
for step in range(5000):

  with tf.GradientTape() as tape:
    hypothesis=tf.sigmoid(tf.matmul(X,w)+b)
    print('hypothesis=',hypothesis)

    #-1*(y*tf.log(hypothesis)+(1-y)*tf.log(1-hypothesis))  # 데이터 1개에 대한것. 데이터수로 원래 나눠줘야해
    cost = -tf.reduce_mean(y*tf.log(hypothesis)+(1-y)*tf.log(1-hypothesis))
    print('cost=',cost)
    grads = tape.gradient(cost,[w,b])
    print('grads(기울기:',grads)

    optimizer.apply_gradients(grads_and_vars=zip(grads,[w,b]))
    print('업데이트 된 w:',w)
    print('업데이트 된 b:',b)
    print('='*100)
```
